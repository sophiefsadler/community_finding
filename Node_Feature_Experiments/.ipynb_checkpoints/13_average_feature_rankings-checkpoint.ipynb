```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
import yaml
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

for alg in ['Louvain', 'Infomap', 'LPA']:
    with open ('rankings/overall_{0}_kmeans.yml'.format(alg.lower()), 'r') as f:
        overall_kmeans = yaml.load(f)
    with open ('rankings/mu_2_{0}_kmeans.yml'.format(alg.lower()), 'r') as f:
        mu_2_kmeans = yaml.load(f)
    with open ('rankings/mu_3_{0}_kmeans.yml'.format(alg.lower()), 'r') as f:
        mu_3_kmeans = yaml.load(f)
    with open ('rankings/mu_4_{0}_kmeans.yml'.format(alg.lower()), 'r') as f:
        mu_4_kmeans = yaml.load(f)

    with open ('rankings/overall_{0}_median.yml'.format(alg.lower()), 'r') as f:
        overall_median = yaml.load(f)
    with open ('rankings/mu_2_{0}_median.yml'.format(alg.lower()), 'r') as f:
        mu_2_median = yaml.load(f)
    with open ('rankings/mu_3_{0}_median.yml'.format(alg.lower()), 'r') as f:
        mu_3_median = yaml.load(f)
    with open ('rankings/mu_4_{0}_median.yml'.format(alg.lower()), 'r') as f:
        mu_4_median = yaml.load(f)

    features = {feat: [] for feat in overall_kmeans.values()}
    for ranking, feat in mu_2_kmeans.items():
        features[feat].append(ranking)
    for ranking, feat in mu_3_kmeans.items():
        features[feat].append(ranking)
    for ranking, feat in mu_4_kmeans.items():
        features[feat].append(ranking)

    kmeans_features_av = {feat: np.mean(rankings) for feat, rankings in features.items()}
    kmeans_sorted_features = sorted(kmeans_features_av.items(), key=lambda item: item[1])

    kmeans_final_ranking = {}
    prev_val = 0
    rank = 1
    for i, feat in enumerate(kmeans_sorted_features):
        if feat[1] == prev_val:
            string_to_add = ', ' + feat[0]
            kmeans_final_ranking[rank] += string_to_add
        else:
            kmeans_final_ranking[i+1] = feat[0]
            rank = i+1
            prev_val = feat[1]

    features = {feat: [] for feat in overall_median.values()}
    for ranking, feat in mu_2_median.items():
        features[feat].append(ranking)
    for ranking, feat in mu_3_median.items():
        features[feat].append(ranking)
    for ranking, feat in mu_4_median.items():
        features[feat].append(ranking)

    median_features_av = {feat: np.mean(rankings) for feat, rankings in features.items()}
    median_sorted_features = sorted(median_features_av.items(), key=lambda item: item[1])

    median_final_ranking = {}
    prev_val = 0
    rank = 1
    for i, feat in enumerate(median_sorted_features):
        if feat[1] == prev_val:
            string_to_add = ', ' + feat[0]
            median_final_ranking[rank] += string_to_add
        else:
            median_final_ranking[i+1] = feat[0]
            rank = i+1
            prev_val = feat[1]

    print('{0}'.format(alg.upper()))
    print('~~~~~~~')
    display_side_by_side(pd.DataFrame(pd.Series(overall_kmeans, name='Overall - K-means')), 
                         pd.DataFrame(pd.Series(kmeans_final_ranking, name='Averaging Over Mu Levels - K-means')),
                         pd.DataFrame(pd.Series(overall_median, name='Overall - Median')), 
                         pd.DataFrame(pd.Series(median_final_ranking, name='Averaging Over Mu Levels - Median')))
    print('\n')
```
