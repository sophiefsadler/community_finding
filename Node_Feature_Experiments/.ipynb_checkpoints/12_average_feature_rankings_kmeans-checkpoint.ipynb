```python
import os
import yaml
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
%matplotlib inline

from pprint import pprint

from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')
```

```python
runs = 100
```

## Feature Ranking for All Graphs

```python
X_train_locs = []
y_train_locs = []

titles = []

for alg in ['Louvain', 'Infomap', 'LPA']:
    X_train_locs.append('../LFR_Graph_Data/1000_Node_Params/{0}_Data/all/node_x_train_kmeans.csv'.format(alg))
    y_train_locs.append('../LFR_Graph_Data/1000_Node_Params/{0}_Data/all/node_y_train_kmeans.csv'.format(alg))
    titles.append(alg)

alg_cols = ['pink', 'purple', 'green']
```

```python
fig = plt.figure(figsize=(15,45))
for exp in range(3):
    
    X_train = pd.read_csv(X_train_locs[exp], index_col=0)
    y_train = pd.read_csv(y_train_locs[exp], index_col=0)
    
    feature_list = list(X_train.columns)
    data = np.array(X_train)
    labels = np.squeeze(np.array(y_train))

    accuracy_scores = []
    feature_importances = {}
    for f in feature_list:
        feature_importances[f] = []
    rows = []

    skf = StratifiedKFold(n_splits=5)
    rf = RandomForestClassifier()

    fold_count = 0

    for run in range(1, runs+1):

        for i in range(5):

            fold_count += 1

            folds = next(skf.split(data, labels), None)

            X_train_fold = data[folds[0], :]
            X_val_fold = data[folds[1], :]
            y_train_fold = labels[folds[0]]
            y_val_fold = labels[folds[1]]

            model = rf.fit(X_train_fold, y_train_fold)
            predictions = rf.predict(X_val_fold)

            accuracy_scores.append(metrics.accuracy_score(y_val_fold, predictions))

            row = {"run": fold_count}
            for j, feature in enumerate(feature_list):
                row[feature] = rf.feature_importances_[j]
            rows.append(row)
    df_importances = pd.DataFrame(rows).set_index("run")
    feature_ranking = df_importances.mean().sort_values(ascending=False)
    ax = fig.add_subplot(1,3,exp+1)
    alg_col = alg_cols[exp]
    feature_ranking[::-1].plot.barh(fontsize=13, figsize=(15,15), color=alg_col, zorder=3)
    ax.set_xlabel("Mean Feature Importance", fontsize=13)
    ax.set_title(titles[exp])
    ax.xaxis.grid()
fig.suptitle('Overall Feature Rankings Using K-means Cutoff for Stable/Unstable', fontsize=16)
fig.tight_layout(pad=2)
fig.subplots_adjust(top=0.9)
```

## Calculating Average Feature Rankings - Louvain

```python
with open ('rankings/overall_louvain_kmeans.yml', 'r') as f:
    overall = yaml.load(f)
with open ('rankings/mu_2_louvain_kmeans.yml', 'r') as f:
    mu_2 = yaml.load(f)
with open ('rankings/mu_3_louvain_kmeans.yml', 'r') as f:
    mu_3 = yaml.load(f)
with open ('rankings/mu_4_louvain_kmeans.yml', 'r') as f:
    mu_4 = yaml.load(f)
    
features = {feat: [] for feat in overall.values()}
for ranking, feat in mu_2.items():
    features[feat].append(ranking)
for ranking, feat in mu_3.items():
    features[feat].append(ranking)
for ranking, feat in mu_4.items():
    features[feat].append(ranking)
    
features_av = {feat: np.mean(rankings) for feat, rankings in features.items()}
sorted_features = sorted(features_av.items(), key=lambda item: item[1])

final_ranking = {}
prev_val = 0
rank = 1
for i, feat in enumerate(sorted_features):
    if feat[1] == prev_val:
        string_to_add = ', ' + feat[0]
        final_ranking[rank] += string_to_add
    else:
        final_ranking[i+1] = feat[0]
        rank = i+1
        prev_val = feat[1]

print('The ranking produced when all graphs are included is:', str(overall))
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('Averaging over the 3 different mu levels, the overall ranking is:', str(final_ranking))
```
