# Community Clusters

In this notebook, we check how many times different numbers of communities were being found by each algorithm on each of the original 20 graphs.

```python
import numpy as np
```

```python
def num_communities(graph_name, algo):
    print('~~~~~~~~~~~~~~~~~~~~~~')
    print(graph_name)
    file_name = '../LFR_Graph_Data/1000_Node_Params/Community_Data/' + algo + '/Runs/' + graph_name + '_runs.npy'
    runs = np.load(file_name)
    comms_freq = {}
    for run_num in range(runs.shape[0]):
        run = runs[run_num, :]
        num_comms = np.max(run)
        if num_comms in comms_freq.keys():
            comms_freq[num_comms] += 1
        else:
            comms_freq[num_comms] = 1
    for i,j in comms_freq.items():
        print('{0} Communities are found {1} times'.format(i,j))
```

## Louvain

```python
for i in [1, 2, 3, 4]:
    for j in [1, 2, 3, 4, 5]:
        graph_name = 'graph_0{0}_mu_0_{1}'.format(j,i)
        num_communities(graph_name, 'Louvain')
```

## LPA

```python
for i in [1, 2, 3, 4]:
    for j in [1, 2, 3, 4, 5]:
        graph_name = 'graph_0{0}_mu_0_{1}'.format(j,i)
        num_communities(graph_name, 'LPA')
```

## Infomap

```python
for i in [1, 2, 3, 4]:
    for j in [1, 2, 3, 4, 5]:
        graph_name = 'graph_0{0}_mu_0_{1}'.format(j,i)
        num_communities(graph_name, 'Infomap')
```
