```python
import pandas as pd
import numpy as np
```

```python
for alg in ['Louvain', 'Infomap', 'LPA']:
    for mu in [2, 3, 4]:
        for graph_no in [1, 2, 3, 4, 5]:
            entrops = pd.read_csv('Community_Data/{0}/Pair_Entropies/graph_0{1}_mu_0_{2}_entropies.csv'.format(alg, graph_no, mu),
                                  index_col=0)
            raw_vals = np.array(entrops['Entropy'])
            print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
            print('Graph {0} At Mu Level 0.{1} with {2}'.format(graph_no, mu, alg))
            print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
            print('Different Community:', (raw_vals < 0.5).sum())
            print('Same Community:', (raw_vals >= 0.5).sum())
```
