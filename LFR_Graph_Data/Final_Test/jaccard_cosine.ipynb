```python
import pandas as pd
import numpy as np
```

```python
same_vals = []
diff_vals = []
for graph in [1, 2, 3, 4, 5]:
    entrops = pd.read_csv('Community_Data/Infomap/Pair_Entropies/graph_0{0}_mu_0_2_entropies.csv'.format(graph), index_col=0)
    feats = pd.read_csv('Community_Data/Infomap/Pair_Features/graph_0{0}_mu_0_2_features.csv'.format(graph), index_col=0)
    for nod_1 in range(200):
        for nod_2 in range(nod_1 + 1, 200):
            entrop = entrops.loc['nodes_{0}_{1}'.format(nod_1, nod_2)]['Entropy']
            if int(entrop) == 0:
                diff_vals.append(feats.loc['nodes_{0}_{1}'.format(nod_1, nod_2)]['Cosine Similarity'])
            elif int(entrop) == 1:
                same_vals.append(feats.loc['nodes_{0}_{1}'.format(nod_1, nod_2)]['Cosine Similarity'])
```
