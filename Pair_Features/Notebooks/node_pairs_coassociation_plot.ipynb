```python
import numpy as np
import networkx as nx
import yaml
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
%matplotlib inline
```

```python
coassociation_matrices = []
```

```python
for i in [1,2,3,4]:
    for j in [1,2,3,4,5]:
        with open ('../lfr_graphs/mu_0_{0}/graph_0{1}/graph_0{1}_mu_0_{0}_coassociation.yml'.format(i,j)) as f:
            c = yaml.load(f, Loader=yaml.Loader)
        coassociation_matrices.append(c)
```

```python
coassociation_values = []
```

```python
for mat in coassociation_matrices:
    mat_vals = [mat[i,j] for i in range(200) for j in range(200) if i < j]
    coassociation_values += mat_vals
```

```python
x = np.array(coassociation_values)
```

```python
plt.hist(x, bins=10)
```

```python
plt.hist(x, bins=25)
```

```python
for mat in coassociation_matrices:
    mat_vals = np.array([mat[i,j] for i in range(200) for j in range(200) if i < j])
    plt.hist(mat_vals, bins=20)
    plt.show()
```
